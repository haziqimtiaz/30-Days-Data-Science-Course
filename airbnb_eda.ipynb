## 3. Data Cleaning

Checking for extra columns that are not needed for our analysis. We can see that there are 5 columns that are not needed for our analysis. We will drop these columns.
- id - This is a unique identifier for each listing. We will not need this column for our analysis.
- name - This is the name of the listing. We will not need this column for our analysis.
- host_id - This is a unique identifier for each host. We will not need this column for our analysis.
- host_name - This is the name of the host. We will not need this column for our analysis.
- last_review - This is the date of the last review. We will not need this column for our analysis.

```python
#dropping the columns that are not needed
df.drop(['id','name','host_id','host_name','last_review'],axis=1,inplace=True)
```

```python
df.head(3)
```

Shifing the columns to make it more readable. 

```python
#shifting the columns
df = df[['room_type','price','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count','availability_365','neighbourhood_group','neighbourhood','latitude','longitude']]
df.head(3)
```

```python
#changing the name of the calculated_host_listings_count column
df.rename(columns={'calculated_host_listings_count':'listings_count'},inplace=True)
```

Checking for missing values to see if we need to impute any values.

```python
#null values
df.isnull().sum()
```

```python
#solve the null values in reviews_per_month column
df['reviews_per_month'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

We checked for missing values and found that there were 10052 missing values in the reviews_per_month column. We will impute these values with 0. This is because if there are no reviews, then the number of reviews per month will be 0.

Now we will check for outliers in the numerical columns. We will use boxplots to check for outliers.

```python
#numerical columns
df.select_dtypes(include=['int64','float64']).columns
```

```python
#checking the outliers in price column
sns.boxplot(y=df['price'])
```

```python
#iqr method
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3-q1
df=df[(df['price']>=q1-1.5*iqr) & (df['price']<=q3+1.5*iqr)]
sns.boxplot(y=df['price'])
```

```python
#minimum_nights column
sns.boxplot(y=df['minimum_nights'])
```

```python
#iqr method
q1 = df['minimum_nights'].quantile(0.25)
q3 = df['minimum_nights'].quantile(0.75)
iqr = q3-q1
df=df[(df['minimum_nights']>=q1-1.5*iqr) & (df['minimum_nights']<=q3+1.5*iqr)]
sns.boxplot(y=df['minimum_nights'])
```

```python
df.shape
```

```python
#availability_365 column
sns.boxplot(y=df['availability_365'])
```

We removed the outliers from the price and minimum_nights columns. We used the IQR method to remove the outliers. Now data is ready for visualization and analysis.
