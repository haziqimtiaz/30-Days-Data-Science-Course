# Data Visualization

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('titanic')
data.head()
```

```python
data.info()
```

```python
#survived plot
data.value_counts('survived').plot(kind='bar')
```

```python
#count plot
sns.countplot(x='survived',data=data)
```

```python
#male and female
sns.countplot(x='survived',hue='sex',data=data)
```

```python
#scatter plot
plt.scatter(data=data,x='age',y='fare')
```

```python
#pclass
sns.scatterplot(data=data,x='age',y='fare',hue='pclass')
```

```python
#bar plot
sns.barplot(data=data,x='age',y='fare',hue='pclass')
```

```python
#pair plot
data.hist()
```
